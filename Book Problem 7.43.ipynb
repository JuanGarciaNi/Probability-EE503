```python
import numpy as np
#import random
import pandas as pd  #pandas for dataframes
import matplotlib.pyplot as plt
```

```python
N=20
Y = np.empty((N,6))

for k in range(1,N+1):
    S1=np.random.choice(np.arange(2), p=[1/2,1/2],size=(6))
    for i in range(1,6+1):
        result1 = (2**i)*np.prod(S1[:i])
        Y[k-1,i-1] = result1
#print('Y =',Y)
for j in range(1,N):
    plt.plot(Y[j],'-o')
plt.ylabel('$Y_n$')
plt.xlabel('$n-1$')
plt.show()
```
